```python
import os
import csv
import json
import pickle
```

```python
airport_file = "../common/airports.csv"

aiport_json = "../../data/aviation-edge/airports.json"
```

```python
def json_load(filepath):
    with open(filepath, "r") as json_file:
        data = json.load(json_file)
    return data

def json_dump(data, filepath, pretty_format = True):
    with open(filepath, 'w') as fw:
        if pretty_format:
            json.dump(data, fw, indent=2, sort_keys=True)
        else:
            json.dump(data, fw)

def pickle_dump(obj, pickle_filepath):
    with open(pickle_filepath, "wb") as f:
        pickle.dump(obj, f, protocol=2)

def pickle_load(pickle_filepath):
    with open(pickle_filepath, "rb") as f:
        obj = pickle.load(f)
    return obj
```

```python
def read_airport_file(filepath):
    airports = set()
    with open(filepath, 'r') as f:
        for line in f:
            airports.add(line.strip())
    return list(airports)
        
```

```python
airport_list = read_airport_file(airport_file)

airport_json_list = json_load(aiport_json)
```

```python
def str_match(name, airport_json_list):
    name1 = name.lower()
    match_list = []
    for idx, obj in enumerate(airport_json_list):
        airport = obj['nameAirport'].lower()
        country = obj['nameCountry'].lower()
        key = airport + " " + country
        
        if name1 in key:
            match_list.append(obj)
            
    return match_list
        
```

```python
match_airports = {}
a_match_airports = {}
for name in airport_list:
    match_list = str_match(name, airport_json_list)
    if len(match_list) == 1:
        match_airports[name] = match_list[0]
    elif len(match_list) > 1:
        a_match_airports[name] = match_list
```
